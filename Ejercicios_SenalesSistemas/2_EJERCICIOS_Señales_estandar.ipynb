**EJERCICIO PROPUESTO 1:** *SEÑAL EXPONENCIAL COMPLEJA*

* Pruebe otros valores de omega y sigma para crear señales con amplitud creciente/decreciente/constante y diferentes frecuencias angulares.

Creamos la función $e^{(\sigma +jω)t}$ en código, la cual usaremos para graficar sus tres casos.

```python
%matplotlib inline
import sympy as sym #importa la librería sympy
sym.init_printing() #se inicializa para mostrar en pantalla

t, sigma, omega = sym.symbols('t sigma omega', real=True)#definir variables simbólicas
s = sigma + 1j*omega  #sobre carga operador para definir parte imaginaria
x = sym.exp(s*t)#función exponencial simbólica desde sympy
x
```

**Creciente**

Para que la señal sea creciente con el tiempo $\sigma$ debe ser mayor a 0.

$\sigma = 0.7$

$ω =4$

Ahora, tomando estos valores específicos para la frecuencia compleja  $s=σ+jω$ , una nueva señal es definida mediante la sustitución de  σ  y  ω con los valores propuestos en el código y su respectiva gráfica.


```python
#creciente -> sigma > 0
y = x.subs({omega: 4, sigma : 0.7}) #omega=4, sigma=0.7
print(r'x(t)=',x)

sym.plot(sym.re(y), (t, 0, 4*sym.pi), ylabel=r'Re{$e^{st}$}',ylim=[-2000,2000])
sym.plot(sym.im(y), (t, 0, 4*sym.pi), ylabel=r'Im{$e^{st}$}',ylim=[-2000,2000]);
```

Como se aprecia en la gráfica, la función $e^{(0.7+4j)t}$ es creciente.

**Decreciente**

Para que la señal sea creciente con el tiempo $\sigma$ es menor a 0.

$\sigma = -0.7$

$ω =4$

Tomando estos valores específicos para la frecuencia compleja  $s=σ+jω$ , una nueva señal es definida mediante la sustitución de  σ  y  ω con los valores propuestos en el código y su respectiva gráfica.


```python
#decreciente sigma < 0
y = x.subs({omega: 4, sigma : -0.7}) #omega=4, sigma=-0.7
print(r'x(t)=',x)

sym.plot(sym.re(y), (t, 0, 2*sym.pi), ylabel=r'Re{$e^{st}$}',ylim=[-1,1])
sym.plot(sym.im(y), (t, 0, 2*sym.pi), ylabel=r'Im{$e^{st}$}',ylim=[-1,1]);
```

Como se aprecia en la gráfica, la función $e^{(-0.7+4j)t}$ es decreciente.

**Constante**

Para que la señal sea constante  $\sigma$ es igual a 0.

$\sigma = 0$

$ω =4$

Tomando estos valores específicos para la frecuencia compleja  $s=σ+jω$ , una nueva señal es definida mediante la sustitución de  σ  y  ω con los valores propuestos en el código y su respectiva gráfica.


```python
#econstante sigma = 0
y = x.subs({omega: 4, sigma : 0}) #omega=4 sigma=0
print(r'x(t)=',x)

sym.plot(sym.re(y), (t, 0, 2*sym.pi), ylabel=r'Re{$e^{st}$}',ylim=[-2,2]) #r' indica al interprete una variable tipo string
sym.plot(sym.im(y), (t, 0, 2*sym.pi), ylabel=r'Im{$e^{st}$}',ylim=[-2,2]);
```

Como se aprecia en la gráfica, la función $e^{(0+4j)t}=e^{4jt}$ es constante.

**EJERCICIO PROPUESTO 2:** *IMPULSO DE DIRAC - FUNCIÓN DELTA δ(t)*

Compruebe la propiedad de selectividad, para un impulso de Dirac desplazado $\delta(t-\tau)$ y probar los resultados modificando el ejemplo.
Definimos la integral de un impulso de Dirac desplazado en $t0$ que está multiplicado por una función $x(t)$.

\begin{equation}
    \int_{-\infty}^{\infty} \delta(t-t0) \cdot x(t)dt = x(t0)
    \end{equation}


```python
#definimos t0
t0 = sym.symbols('t0', real=True)
x = sym.Function('x')(t) #x(t)
delta_T = sym.DiracDelta(t-t0) #delta(t-t0)
delta_T
sym.integrate(delta_T*x, (t, -sym.oo, sym.oo))
```

Se comprueba la propiedad de selectividad en el código, ahora podemos darle un valor a $t0=\pi$, y usar una función $x(t)=cos(t)$.
\begin{equation}
    \int_{-\infty}^{\infty} \delta(t-\pi) \cdot cos(t)dt = cos(\pi)=-1
    \end{equation}

```python
#definir t y t0
t,t0 = sym.symbols('t t0', real=True)

#asigno un valor a t0 para comprobar selectividad
t0=sym.pi

x = sym.cos(t)
delta_T = sym.DiracDelta(t-t0) # delta(t-t0)
delta_T
sym.integrate(delta_T*x, (t, -sym.oo, sym.oo))
```

Podemos también comprobar la propiedad de escalamiento
\begin{equation}
    \int_{-\infty}^{\infty}\delta(at-t0) \cdot x(t)dt = \frac{1}{|a|} \cdot x(\frac{t0}{a})
    \end{equation}

\begin{equation}
    \int_{-\infty}^{\infty}\delta(2t-5) \cdot cos(t)dt = \frac{cos(5/2)}{2}
    \end{equation}

```python
#definir t a y t0
t, a, t0 = sym.symbols('t a t0', real=True)

#asigno valores para comprobar selectividad y escalamiento
a=2
t0=5

g = sym.cos(t)
delta_T = sym.DiracDelta(a*t-t0) # delta(at-t0)
delta_T
sym.integrate(delta_T*g, (t, -sym.oo, sym.oo))
```

**EJERCICIO PROPUESTO 3:** *SEÑAL PULSO RECTANGULAR*

Use la función $\text{rect}(t)$ para construir una función $\cos(\omega t)$ con $\omega=2$, la cual se enciende en $t=-\frac{1}{2}$ y se apaga en $t=+\frac{1}{2}$.

Definimos las funciónes; pulso rectangular desde el escalón y $cos(ωt)$.

```python
class rect(sym.Function):

    @classmethod #decorador objeto
    def eval(cls, arg):
        return sym.Heaviside(arg + sym.S.Half) - sym.Heaviside(arg - sym.S.Half)

omega = sym.symbols('$\omega$', real=True)
x = sym.cos(omega*t)
```

Creamos una nueva función $y$, que es el producto de la función rectangular y $cos(ωt)$, definimos para la frecuencia $ω=2$ y graficamos esta nueva función.

```python
y = x*rect(t)
sym.plot(y.subs(omega,2), (t, -1, 1), ylim=[-0.2, 1.2], ylabel=r'y(t)=cos(ωt)rect$(t)$');
```

Como se mira en la gráfica al multiplicar una función, ($cos(2t)$ en nuestro caso) la función solo tendra amplitud en los puntos donde la función rectangular sea diferente a 0.

**EJERCICIO PROPUESTO 4:** *FUNCIÓN SIGNO*

Generar un rectificador de onda completa utilizando la función $sgn(t)$.


Definimos las funciones; $sgn(t)$ y $y(t)=sen(2t)$

```python
sgn = sym.sign(t)
sgn
y =sym.sin(2*t)
```

Al graficar $y(t)=sen(2t)$ obtenemos la siguiente gráfica cuya amplitud toma valores negativos y positivos.

```python
sym.plot(y, (t, -3.5,3.5), ylim=[-1.2, 1.2], ylabel=r'sen$(2t)$');
```

Para modelar un rectificador de onda completa con esta función, utilizamos la función $sgn(t)$ para que los valores en amplitud sean todos positivos, es decir que la función signo me representa el valor absoluto de la señal $y(t)$.

\begin{equation}
|x(t)| = x(t) \cdot \text{sgn}(x(t))
\end{equation}
\begin{equation}
|sen(2t)| = sen(2t) \cdot \text{sgn}(sen(2t))
\end{equation}

En código y con su gráfica queda así:

```python
x = y*sym.sign(sym.sin(2*t))
sym.plot(x, (t, -3.5,3.5), ylim=[-1.2, 1.2], ylabel=r'|sen(2t)|$');
```
